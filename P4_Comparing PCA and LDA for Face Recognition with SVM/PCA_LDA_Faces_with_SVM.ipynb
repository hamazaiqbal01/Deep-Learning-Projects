#Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Load Faces dataset
faces_data = fetch_olivetti_faces(shuffle=True, random_state=22)
X = faces_data.data
y = faces_data.target
```

# Following steps should be performed

# Steps:
- Data Split
- Feature Scaling
- Models ready
- Transform Scaled model into pca or lda

- Predictions
- Accuracy

#Step: 1

#Splitting Data

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

#Step: 2

# features scaling

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# PCA models definition

```python
pca = PCA(n_components=100)
```

# LDA models definition

```python
lda = LinearDiscriminantAnalysis(n_components=10)
```

#Step: 3

# Transform training and testing data using PCA

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

# Transform training and testing data using LDA

```python
X_train_lda = lda.fit_transform(X_train_scaled, y_train)
X_test_lda = lda.transform(X_test_scaled)
```

#Step: 4

# Train a SVM classifier on PCA-transformed data

```python
svm_pca = SVC(kernel='linear')
svm_pca.fit(X_train_pca, y_train)
```

# Train a SVM classifier on LDA-transformed data


```python
svm_lda = SVC(kernel='linear')
svm_lda.fit(X_train_lda, y_train)
```

#Step: 5

# predictions on the test set for PCA and LDA models


```python
y_pred_pca = svm_pca.predict(X_test_pca)
y_pred_lda = svm_lda.predict(X_test_lda)
```

# accuracy for PCA and LDA

```python
accuracy_pca = accuracy_score(y_test, y_pred_pca)
accuracy_lda = accuracy_score(y_test, y_pred_lda)
```

#Accuracy

```python
print("Accuracy PCA:", accuracy_pca)
print("Accuracy LDA:", accuracy_lda)
```

# PCA and LDA Visualization


```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
for i, target_name in enumerate(faces_data.target):
    plt.scatter(X_test_pca[y_test == i, 0], X_test_pca[y_test == i, 1], label=target_name)
plt.title('PCA Transformation of Face Data')
plt.xlabel('PC')
plt.ylabel('PC')
plt.legend()

plt.subplot(1, 2, 2)
#for i, target_name in enumerate(faces_data.target):
for i in range(10):
    plt.scatter(X_test_lda[y_test == i, 0], X_test_lda[y_test == i, 1], label=target_name)
plt.title('LDA Transformation of Face Data')
plt.xlabel('LD')
plt.ylabel('LD')
plt.legend()

plt.tight_layout()
plt.show()
```
