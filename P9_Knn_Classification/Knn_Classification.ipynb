```python
img=Image.open(image_path)
plt.imshow(img)
```

*converting to gray scale image*

```python
gray_image = img.convert("L")
plt.imshow(gray_image,cmap="gray")
```

Negtive Image

```python
# Get the pixel data as a NumPy array
pixel_data = np.array(gray_image)

# Calculate the negative of the image
negative_image = 255 - pixel_data

# Create a new Image object from the negative pixel data
negative_pil_image = Image.fromarray(negative_image)

# Save or display the negative image
plt.imshow(negative_pil_image,cmap="gray")
```

*Enhanced the image*

```python
# Enhance contrast using histogram equalization
sharpened_image = negative_pil_image.filter(ImageFilter.EDGE_ENHANCE)
gray_image=sharpened_image
#plt.imshow(enhanced_image,cmap="gray")
```

*Reshaping the gray scale image*

```python
reshaped_gray_image = gray_image.resize((8, 8))
plt.imshow(reshaped_gray_image,cmap="gray")
```

*Flatten the Image*

```python
reshape_np_array=np.array(reshaped_gray_image)
flat_image=reshape_np_array.flatten()
```

```python
print(flat_image.shape)
print(flat_image.size)
```

```python
test=flat_image
```

```python
test_reshape=test.reshape(1,-1)
```

```python
Loaded_model.predict(test_reshape)
```

#Steps:

>>Importing Libraries and Loading Model:

The code starts by importing necessary libraries, including joblib for loading the trained model.
It loads a KNN classification model using the joblib.load() function and assigns it to the variable Loaded_model.

>>Pre-Processing Steps:

The code uses the PIL library (Python Imaging Library) for image manipulation.
The image is loaded from the specified path using Image.open() and displayed using plt.imshow().

>>Converting to Gray Scale:

The image is converted to grayscale using the .convert("L") method, which retains only the luminance (brightness) information and discards color information.
The grayscale image is displayed using plt.imshow() with the "gray" colormap.

>>Negative Image:

The pixel data of the grayscale image is converted to a NumPy array.
The negative of the image is calculated by subtracting each pixel value from 255 (inverting the grayscale values).
The negative image is created using Image.fromarray() and displayed using plt.imshow() with the "gray" colormap.

>>Enhanced Image:

The negative image is filtered using the EDGE_ENHANCE filter from the ImageFilter module, creating a sharpened effect.
The sharpened image is assigned to gray_image.

>>Reshaping the Gray Scale Image:

The sharpened image is resized to a size of 8x8 using .resize((8, 8)).
The resized image is displayed using plt.imshow() with the "gray" colormap.

>>Flattening the Image:

The 8x8 image is flattened into a 1D array using the .flatten() method, resulting in a 64-element array.
The shape and size of the flattened array are printed.
Making a Prediction:The flattened array is reshaped to match the input format expected by the loaded KNN model (test_reshape).
The Loaded_model.predict() function is used to predict the digit class based on the processed image data.
>>Conclusion

In this code, we explored a process for classifying hand-drawn digits using a pre-trained K-Nearest Neighbors (KNN) classification model. The process involved several image pre-processing steps to enhance the quality and features of the input image before making a prediction. The steps included converting the image to grayscale, applying negative transformation, enhancing edges, resizing the image to a standard size, and flattening the image for input into the model.
